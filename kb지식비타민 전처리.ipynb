## 요약 추출  
- 맨처음나오는 \n \n\uf071 ~ < 요   약 >

```python
data['check'] = data['txt'].apply(lambda x:1 if '< 요   약 >' in x else 0)
```

```python
data[data['check']==1]
```

```python
data[data['check']==1].index
```

```python
summary_list = []
total_list = []

for i in data[data['check']==1].index:
    tmp = data.loc[i,'txt']
    start_pos = tmp.find('\n \n\uf071')+5 # '\n \n\uf071'
    end_pos = tmp.find('< 요   약 >')
    value = tmp[start_pos:end_pos]
    total = tmp[end_pos:]
    
    summary_list.append(value)
    total_list.append(total)
```

```python
summary_list[1] ## 유독 이상함
```

```python
## idx = 159
tmp = data.loc[159,'txt']
start_pos = tmp.find('< 요   약 >')+13
end_pos = tmp.find('< 요   약 >',start_pos+1)
value = tmp[start_pos:end_pos]
total = tmp[end_pos:]
value
```

```python
summary_list[1] = value # 값 대체
total_list[1] = total
```

```python
summary_list
```

### 요약, 전체 내용(요약 제외) 데이터프레임에 넣기

```python
j=0
for i in range(len(data)):
    if data.loc[i,'check']==1:
        data.loc[i,'summary'] = summary_list[j]
        data.loc[i,'total'] = total_list[j]
        j=j+1
    else:
        data.loc[i,'summary'] = 0
        data.loc[i,'total'] = 0
```

## 줄바꿈 제거

```python
import re
```

```python
for i in range(len(data)):
    tmp1 = re.sub("\n", "", str(data.loc[i,'summary']))
    tmp2 = re.sub("\n", "", str(data.loc[i,'total']))
    data.loc[i,'summary']=tmp1
    data.loc[i,'total']=tmp2
```

## 특수문자 제거
https://codechacha.com/ko/python-remove-special-letters/

```python
for i in range(len(data)):
    tmp1 = re.sub(r"[^\uAC00-\uD7A30-9a-zA-Z\s]", "", str(data.loc[i,'summary']))
    tmp2 = re.sub(r"[^\uAC00-\uD7A30-9a-zA-Z\s]", "", str(data.loc[i,'total']))
    data.loc[i,'summary']=tmp1
    data.loc[i,'total']=tmp2
```

## 자주 나오는 단어 제거  
- 단어 개수 센 다음 제거
---
- [그림]은 지워야 함  
- < 요   약 > 있는 애들 지우기  
- 공백 2번 이상인 애들 지우기 : 앞에서 특수문자하면서 지워짐

```python
for i in range(len(data)):
    tmp1 = re.sub("\[그림", "", str(data.loc[i,'summary']))
    tmp2 = re.sub("\[그림", "", str(data.loc[i,'total']))
    data.loc[i,'summary']=tmp1
    data.loc[i,'total']=tmp2
```
