```python
import pandas as pd 

df = pd.read_csv('KB골든라이프X_통합.csv',encoding='cp949')
df.head()
```

# 결측치 제거

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['text'], how='any', axis=0)
df = df.reset_index(drop=True)
```

```python
df['text'][0]
```

# 문장 분리  
- kiwipeipy 이용

```python
!pip install kiwipiepy
```

```python
from kiwipiepy import Kiwi
kiwi = Kiwi()
```

```python
txt_list = []

for i in range(len(df)):
    print('=============',i,'번째=============')
    content = df['text'][i]
    content_split = kiwi.split_into_sents(content)
    
    for j in range(len(content_split)):
        txt_list.append(content_split[j][0])
```

```python
txt_list
```

```python
txt_df = pd.DataFrame()
txt_df['txt'] = txt_list
```

```python
txt_df
```

# 중복 문장 제거

```python
txt_df.duplicated(keep=False).sum()
```

```python
txt_df[txt_df.duplicated(keep=False)]
```

```python
txt_df = txt_df.drop_duplicates()
txt_df = txt_df.reset_index(drop=True)
```

# 길이 짧은 문장 제거

```python
_len = []
for i in range(len(txt_df)):
    _len.append(len(txt_df.loc[i,'txt']))
```

```python
txt_df['len'] = _len
```

```python
txt_df['len'].describe() ## 문장 길이
```

문장 길이가 10 이상 350 이하가 되는 문장만 선택

```python
txt_df = txt_df[(10<=txt_df['len']) & (txt_df['len']<=350)]
txt_df = txt_df.reset_index(drop=True)
```
