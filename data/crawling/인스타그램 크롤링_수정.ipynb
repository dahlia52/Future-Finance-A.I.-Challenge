# 1. url 추출

## url 추출 함수

```python
def find_url(driver):
    # url 추출
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    num1 = soup.select('a.x1i10hfl')
    flag = 0
    global extracted_value
    
    for num in num1:
        ##====================== 수정 필요 ===================================
        pattern = r'/p/([a-zA-Z0-9-_]+)/liked_by/' ## 형식 수정해야 함
        ##===================================================================
        text = str(num)
        result = re.search(pattern, text)
        
        if result:
            extracted_value = result.group(1)
            flag =  1
            print('url 추출 완료')
            break
    
    if flag == 0:
        print('url 추출 실패')
        return 0
    else:
        return extracted_value
```

## 로그인 함수

```python
def login(id, pw):
    # 로그인 페이지로 이동
    driver.get("https://www.instagram.com/accounts/login/")
    time.sleep(1)
    
    # id와 pw를 입력하는 창의 요소 정보 획득
    input = driver.find_elements_by_tag_name("input")

    # 아이디를 입력
    input[0].send_keys(id)

    # 비밀번호 입력
    input[1].send_keys(pw)

    # 엔터
    input[1].send_keys(Keys.RETURN)
    time.sleep(5)
```

## 정보 추출 함수

```python
# 함수 정의: 본문 내용, 작성일자, 좋아요 수, 해시태그 가져오기
import re
from bs4 import BeautifulSoup

def get_content(driver):
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    
    # 본문 내용
    try:
        content = soup.select('div._a9zs')[0].text
    except:
        content = ''
            
    # 해시태그
    tags = re.findall(r'#[^\s#,\\]+', content)
    
    # 작성일자
    date = soup.select('time._aaqe')[0]['datetime'][:10]
    
    # 좋아요
    try:
        like = soup.select('section._ae5m._ae5n._ae5o')[0].findAll('span')[-1].text
    except:
        like = 0

    return content, date, like, tags
```

## main

```python
# 아이디, 비밀번호 설정
# id = ""
# pw = ""

# 크롬 웹드라이버 실행
driver = webdriver.Chrome(r"C:\chromedriver.exe")

# 로그인
login(id,pw)

# 국민은행 접속
url = "https://www.instagram.com/kbkookminbank/"
driver.get(url)
time.sleep(3)
    
# 첫번째 게시물 클릭
first_img_css = "div._aagu"
driver.find_element_by_css_selector(first_img_css).click()
time.sleep(1)

# url 추출 리스트
no_list = []
content_list = []
date_list = []
like_list = []
tags_list = []

n = 1254 # 게시물 개수
#n = 10
for i in range(1,n):
    print("================",i,"번째================")
    
    ## 글 추출
    content,date,like,tags = get_content(driver)
    content_list.append(content)
    date_list.append(date)
    like_list.append(like)
    tags_list.append(tags)
    no_list.append(find_url(driver))
    
    ## 다음 버튼 클릭
    driver.find_element_by_css_selector('._aaqg ._abl-').click()
```

```python
import pandas as pd

results = pd.DataFrame()
results['date'] = date_list
results['content'] = content_list
results['tags'] = tags_list
results['like'] = like_list

## url 추출 -> no_list가지고 url 형식 만들기
url_list = []

for num in no_list:
    txt = "https://www.instagram.com/p/"+str(num)
    url_list.append(txt)

results['url'] = url_list
```
