```python
import fitz
import pandas as pd
```

```python
issue_df = pd.read_csv('invest_issue_new2.csv')
issue_df.head()
```

```python
pdf_title = issue_df['pdf title'] # 파일명

txt_list = []
title_list = []
    
for i in range(len(pdf_title)):   
    try:
        print('==================')
        print(str(i)+'번째')
    
        if pdf_title[i]=='없음':
            continue
        
        path = 'C:\\이은효\\2023-2\\2. kb 공모전\\1. 데이터\\1. KB종합자산관리\\3. 이슈리포트\\'+ pdf_title[i]
        doc = fitz.open(path)

        total_txt = ''
        title_list.append(pdf_title[i])

        for page in doc:
            text = page.get_text()
            total_txt += text

        txt_list.append(total_txt)
        print('완료')
        
    except:
        pass
```

```python
pdf_txt_df = pd.DataFrame({"pdf_title":title_list,
                           "txt": txt_list})
```

```python
pdf_txt_df[pdf_txt_df['txt']=='']
```

```python
pdf_txt_df.to_csv('invest_issue_pdf_extract.csv')
```

## 최종 데이터

```python
issue_df_new  = issue_df.copy()
```

```python
issue_df_new2 = pd.merge(issue_df_new, pdf_txt_df, left_on='pdf title', right_on='pdf_title')
```

```python
issue_df_new2.drop(['Unnamed: 0','pdf_title'], axis=1, inplace=True)
```
